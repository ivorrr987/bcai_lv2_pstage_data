```python
import os
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mor = pd.read_csv("csv_data/모로코 요리.csv")
som = pd.read_csv("csv_data/소말리아 요리.csv")
```

```python
mor
```

```python
som
```

```python
temp1 = pd.concat([mor, som]) # temp1 = morocco + somalia
temp1
```

```python
labels = sorted(temp1['label'].unique())
counts = temp1['label'].value_counts(sort=False)
print(labels)
print(counts)
```

```python
fig, ax = plt.subplots(figsize=(12,7))
plt.xticks(range(0,17))
ax.bar(labels, counts)
```

```python
counts = temp1['label'].value_counts(sort=False, normalize=True)
```

```python
fig, ax = plt.subplots(figsize=(12,7))
plt.xticks(range(0,17))
ax.bar(labels, counts)
```

```python
total = pd.read_csv("total.csv")
total_labels = sorted(total['label'].unique())[1:]
total = pd.Series(total['label'])
print(total)
total_counts = total.value_counts(sort=False, dropna=True)
print(total_labels)
print(total_counts)
```

```python
fig, ax = plt.subplots(figsize=(12,7))
plt.xticks(range(0,17))
ax.bar(total_labels, total_counts)
```
