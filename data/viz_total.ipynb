```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
total = pd.read_csv("0414_2.csv")
total
```

```python
total_not_nan = total[total['label'].isnull() != True]
total_not_nan
```

```python
labels = sorted(total_not_nan['label'].unique())
print(labels)
counts = total_not_nan['label'].value_counts()
print(counts)
```

```python
fig, ax = plt.subplots(figsize=(12,7))
ax.bar(labels, counts)
```

```python
counts = total_not_nan['label'].value_counts(sort=True, normalize=True)
```

```python
fig, ax = plt.subplots(figsize=(12,7))
ax.bar(labels, counts)
```
